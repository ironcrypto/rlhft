```python
import pandas as pd
import numpy as np
import os 
import re
path="/home/mlqin/qml/HFT_06_17/data/GALAUSDT/train"
list_path=os.listdir(path)
def sort_list(lst: list):
    convert = lambda text: int(text) if text.isdigit() else text
    alphanum_key = lambda key: [convert(c) for c in re.split("([0-9]+)", key)]
    lst.sort(key=alphanum_key)
sort_list(list_path)
```

```python
return_rate_list=[]
for leaf_path in (list_path):
    df=pd.read_feather(os.path.join(path,leaf_path))
    return_rate=df.iloc[-1]["bid1_price"]/df.iloc[0]["ask1_price"]-1
    return_rate_list.append(return_rate)
return_rate_list
```
